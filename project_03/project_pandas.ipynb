```python
#11. Plot a histogram showcasing the distribution of movie ratings.
data['Rating'].plot(kind='hist', bins=10)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# 12. Calculate the correlation coefficient between movie ratings and runtimes.
correlation = data['Rating'].corr(data['Runtime (Minutes)'])
print("Correlation Coefficient:",correlation)
```

```python
#13. Identify the 3 most frequent actors in the dataset.
actor_counts = data['Actors'].str.split(', ').explode().value_counts().head(3)
print(actor_counts)
```

```python
# 14. Is there any visible relationship between box office earnings and movie ratings?
plt.scatter(data['Revenue (Millions)'], data['Rating'])
plt.xlabel('Revenue (Millions)')
plt.ylabel('Rating')
plt.show()
```

```python
# 20. Plot the average movie ratings over the years to identify trends.
avg_ratings_by_year = data.groupby('Year')['Rating'].mean()
avg_ratings_by_year.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```

```python

genre = "Horror"  

#Filter by genre
genre_data = data[data['Genre'].str.contains(genre)]

#Sort by ratings and votes in descending order
sorted_genre_data = genre_data.sort_values(by=['Rating', 'Votes', 'Revenue (Millions)'], ascending=False)

#Select top 10 movies
top_10_movies = sorted_genre_data.head(10)

print(top_10_movies[['Title', 'Rating', 'Votes', 'Revenue (Millions)','Director']])
```
